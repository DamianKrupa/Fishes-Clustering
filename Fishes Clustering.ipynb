```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
raw_data = pd.read_csv('Fish.csv')
data = raw_data.copy()
```

## Unscaled data 

```python
# create a variable which will contain the data for the clustering
x = data.iloc[:,1:].copy()
# create a k-means object with 2 clusters
kmeans = KMeans(7)
# fit the data
kmeans.fit(x)
```

```python
# create a copy of data, so we can see the clusters next to the original data
clusters = data.iloc[:,1:].copy()
# predict the cluster for each observation
clusters['cluster_pred']=kmeans.fit_predict(x)
```

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
 
fig = plt.figure(figsize = (15, 10))
ax = plt.axes(projection ="3d")
 
ax.scatter3D(clusters['Width'], clusters['Weight'], clusters['Length3'], c= clusters['cluster_pred'], cmap = "rainbow")
plt.title("simple 3D scatter plot")
plt.show()
```

## Standardize the variables

```python
# import some preprocessing module
from sklearn import preprocessing
# scale the data for better results
x_scaled = preprocessing.scale(data.iloc[:,1:])
x_scaled
```

## Scaled data

```python
# create a k-means object with 2 clusters
kmeans_scaled = KMeans(7)
# fit the data
kmeans_scaled.fit(x_scaled)
```

```python
# create a copy of data, so we can see the clusters next to the original data
clusters_scaled = data.copy()
# predict the cluster for each observation
clusters_scaled['cluster_pred']=kmeans_scaled.fit_predict(x_scaled)
```

```python
fig = plt.figure(figsize = (15, 10))
ax = plt.axes(projection ="3d")
 
ax.scatter3D(clusters_scaled['Width'], clusters_scaled['Weight'], clusters_scaled['Length3'], c= clusters['cluster_pred'], cmap = "rainbow")
plt.title("simple 3D scatter plot")
plt.show()
```

```python

fig = plt.figure(figsize = (15, 10))
ax = plt.axes(projection ="3d")
 
ax.scatter3D(clusters_scaled['Length3'], clusters_scaled['Weight'], clusters_scaled['Height'], c= clusters['cluster_pred'], cmap = "rainbow")
plt.title("simple 3D scatter plot")
plt.show()
```

## Take Advantege of the Elbow Method

```python
wcss = []
# 'cl_num' is a that keeps track the highest number of clusters we want to use the WCSS method for. We have it set at 10 right now, but it is completely arbitrary.
cl_num = 10
for i in range (1,cl_num):
    kmeans= KMeans(i)
    kmeans.fit(x_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
wcss
```

```python
number_clusters = range(1,cl_num)
plt.plot(number_clusters, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```
